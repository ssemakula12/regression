```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("/home/mosiko/regression/datasets/Data1.csv")
data
```

```python
'''As you can see in the picture above, we assume that the number of views depends on the number of blogs. Therefore,

Independent Variable -> x-axis -> Number of blogs

Dependent Variable -> y-axis -> Total views'''
```

```python
#we plot the data using SCATTER DIAGRAM

plt.scatter(data.Blogs,data.Views,color="green")
plt.xlabel("number of Blogs")
plt.ylabel("Number of views")
plt.title("A SCATTER DIAGRAM SHOWING THE RELATIONSHIP BETWEEN THE BLOGS AND VIEWS ")
```

```python
#In this model, we need to find the best-fit line for this graph. To do that, we need to train this model. The model can automatically find the best-fit line.

#Views and Blogs data are stored in the Pandas arrays as series. You can see it like this:
```

```python
data.Blogs
```

```python
data.Views
```

```python
#But before we move on, we need to store them as NumPy arrays.

x=np.array([data.Blogs.values])
x
```

```python
y=np.array([data.Views.values])
y
```

```python
y.shape
#(1,5) means one sample and 5 target values hence needs to be reshaped 
```

```python
x.shape
#(1,5) means one sample and 5 features hence needs to be reshaped
```

```python
#Now, it is done! Let’s proceed.
```

```python
model=LinearRegression()
model.fit(x.reshape((-1,1)),y.ravel())
```

```python
#y.ravel() means you are reshaping y variable to make it one target variable; (n_samples,)
```

```python
#model = LinearRegression()

"""The above definition creates an instance of the LinearRegression class and assigns it to the variable model.
At this point, the model is initialized, but it hasn’t been trained on any data yet."""
```

```python
#model.fit(x.reshape((-1,1)), y.ravel())

'''Here, we’re using the fit method to train the linear regression model. 
The x and y are presumably your input features and target values, respectively.'''
```

```python
#x.reshape((-1,1))

'''As you know, our x is a 2-dimensional array.
-1 means that there should automatic generation of rows basing on the available features
1 means all the features should be in one column
The target variable is expected in 1D, as our y.'''
#by doing all the above the model is trained!!!
```

```python
#If blog_6 has 14 blogs, let’s predict how many views there would be.
test_x=np.array([14]).reshape((-1,1))
test_x
```

```python
predictable_views=model.predict(test_x)
predictable_views
```

```python
#As you can see, we can predict around 418 views!!!
```

```python
'''Before we end this, try to plot the best-fit line of this graph. 
To do that, we need to find the gradient and intercept. 
For that, we use the polyfit function in the NumPy library.'''
```

```python
m,c=np.polyfit(data.Blogs,data.Views,1)
m,c
```

```python
#fit the line y=mx+c;
y_onedimension=m*data.Blogs+c
y_onedimension
```

```python
#plotting A SCATTER DIAGRAM
plt.scatter(data.Blogs,data.Views,color="green")
plt.xlabel("Number of blogs")
plt.ylabel("Number of Views")
plt.title("SCATTER PLOT SHOWING RELATIONSHIP BETWEEN BLOGS AND VIEWS WITH A FIT LINE")
plt.plot(data.Blogs,y_onedimension,color="red",label=f'line: y={m}data.Blogs+{c}')
```
